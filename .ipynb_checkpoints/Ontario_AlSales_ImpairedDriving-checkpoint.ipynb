```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
data_path = os.path.join('Resources', 'canada_population.csv')
ontario = pd.read_csv(data_path)
ontario = ontario[["When", "Ontario"]]
ontario["REF_DATE"] = ontario["When"].str[-4:]
ontario = ontario[["Ontario", "REF_DATE"]]
ontario["REF_DATE"] = ontario["REF_DATE"].astype(int)
ontario = ontario.drop_duplicates(subset=["REF_DATE"], keep='last')
ontario = ontario.loc[(ontario["REF_DATE"] >= 1998)&
                     (ontario["REF_DATE"] <= 2013), :]
ontario
```

```python
data_path = os.path.join('Resources', 'ontario_impaired_driving.csv')
data = pd.read_csv(data_path)
data_clean1 = data.loc[(data["Violations"]=="Total impaired driving [910]") &
                       (data["UOM"] == "Rate") &
                       (data["Statistics"]=="Rate per 100,000 population"),:]
impaired_driving_on = data_clean1[["REF_DATE", "GEO", "Violations", "SCALAR_FACTOR", "VALUE"]]
impaired_driving_on = impaired_driving_on.loc[impaired_driving_on["REF_DATE"] <= 2013, :]
impaired_driving_on
merged_driving = pd.merge(impaired_driving_on, ontario, on="REF_DATE")
merged_driving["Impaired Driving per 100,000"] = (merged_driving["VALUE"]/merged_driving["Ontario"]) * 100000
merged_driving
```

```python
data_path = os.path.join('Resources', 'alcohol_sales_89_18.csv')
data2 = pd.read_csv(data_path)
data_clean1 = data2.loc[(data2["GEO"]=="Ontario") & (data2["Type of sales"]=="Total sales") &
                        (data2["Type of beverage"]=="Total alcoholic beverages"),:]
alcohol_value = data_clean1[["REF_DATE", "GEO", "Type of beverage", "VALUE"]]
alcohol_value = alcohol_value.loc[(alcohol_value["REF_DATE"] >= 1998), :]
merged_alcohol = pd.merge(alcohol_value, ontario, on="REF_DATE")
merged_alcohol["Alcoholic Beverages per 100,000"] = (merged_alcohol["VALUE"]/merged_alcohol["Ontario"]) * 100000
merged_alcohol
```

```python
complete = pd.merge(merged_driving, merged_alcohol, on="REF_DATE")
complete = complete[["REF_DATE", "Impaired Driving per 100,000", "Alcoholic Beverages per 100,000"]]
complete = complete.rename(columns={
    "REF_DATE": "Date",
})
complete
```

```python
x_axis = complete["Alcoholic Beverages per 100,000"]
y_axis = complete["Impaired Driving per 100,000"]
plt.scatter(x_axis, y_axis, marker="o")
plt.title("Alcoholic Beverages Sold v Impaired Driving Incidents in Ontario (1998-2013)")
plt.xlabel("Alcoholic Beverages per 100,000")
plt.ylabel("Impaired Driving per 100,000")
plt.tight_layout()
plt.show()
```
