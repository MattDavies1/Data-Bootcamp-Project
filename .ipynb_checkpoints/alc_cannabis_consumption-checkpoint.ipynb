```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
alc_df = pd.read_csv("Resources/heavy_drinking_data.csv")
alc_df = alc_df.rename(columns={"REF_DATE":"Year","GEO":"Country","VALUE":"Number of People"})
alc_df= alc_df[["Year","Country","Age group","Indicators","Characteristics","Number of People"]]
```

```python
clean_alc_df = alc_df.loc[alc_df["Characteristics"]=="Number of persons",:]
clean_alc_df
```

```python
reduced_alc_df = clean_alc_df.loc[clean_alc_df["Age group"]!="Total, 12 years and over",:].drop(columns=["Characteristics"])
reduced_alc_df
```

```python
groupby_df = reduced_alc_df.groupby(["Year","Age group"]).sum()
test = groupby_df.reset_index()
test
```

```python
teens = test.loc[test["Age group"]== "12 to 17 years"]
young_adults = test.loc[test["Age group"]=="18 to 34 years"]
mid_adults = test.loc[test["Age group"]=="35 to 49 years"]
older_adults= test.loc[test["Age group"]=="50 to 64 years"]
elderly = test.loc[test["Age group"]=="65 years and over"]
w=0.15
x=["2015","2016","2017","2018"]
bar1=np.arange(len(x))
bar2 = [i+w for i in bar1]
bar3 = [i+w for i in bar2]
bar4 = [i+w for i in bar3]
bar5 = [i+w for i in bar4]
```

```python
plt.bar(bar1,teens["Number of People"],w,label="12 to 17 years")
plt.bar(bar2,young_adults["Number of People"],w,label="18 to 34 years")
plt.bar(bar3,mid_adults["Number of People"],w,label="35 to 49 years")
plt.bar(bar4,older_adults["Number of People"],w,label="50 to 64 years")
plt.bar(bar5,elderly["Number of People"],w,label="65+ years")
plt.legend(loc="best")
plt.xticks(bar1+w,x)
plt.ylabel("Number of People Who Drink (millions)")
plt.xlabel("Years")
plt.title("Alcoholic Drinkers by Age Groups")
plt.show()
```

```python
cannibis_df = pd.read_csv("Resources/cannabis_usage_1961_2018.csv")
cannibis_df = cannibis_df.rename(columns={"REF_DATE":"Year","VALUE":"Value"})
cannibis_df.columns
```

```python
clean_can_df = cannibis_df[["Year","Prevalence","Age group","Frequency of use","Value"]]
clean_can_df
```

```python
reduced_can_df = clean_can_df.loc[(clean_can_df["Year"]>=2015)&
                                 (clean_can_df["Prevalence"]=="Consuming population of cannabis")&
                                 (clean_can_df["Age group"]!="65 years and over")&
                                 (clean_can_df["Age group"]!="15 years and over")&
                                 (clean_can_df["Frequency of use"]=="Total, all frequencies"),: ]
```

```python
reduced_can_df
```

```python
can_groupby_df = reduced_can_df.groupby(["Year","Age group"]).sum()
test = can_groupby_df.reset_index()
test
```

```python
teen = test.loc[(test["Age group"]!= "25 to 44 years") & (test["Age group"]!= "18 to 24 years") & (test["Age group"]!= "45 to 64 years"),: ]
young_adults = test.loc[test["Age group"]=="18 to 24 years"]
mid_adults = test.loc[test["Age group"]=="25 to 44 years"]
older_adults= test.loc[test["Age group"]=="45 to 64 years"]
w=0.2
x=["2015","2016","2017","2018"]
bar1=np.arange(len(x))
bar2 = [i+w for i in bar1]
bar3 = [i+w for i in bar2]
bar4 = [i+w for i in bar3]
```

```python
plt.bar(bar1,teen["Value"],w,label="15 to 17 years")
plt.bar(bar2,young_adults["Value"],w,label="18 to 24 years")
plt.bar(bar3,mid_adults["Value"],w,label="25 to 44 years")
plt.bar(bar4,older_adults["Value"],w,label="45 to 64+ years")
plt.legend(loc="best")
plt.xticks(bar1+w,x)
plt.ylabel("Number of People Who Consume Cannabis(millions)")
plt.xlabel("Years")
plt.title("Cannabis Consumers by Age Groups")
plt.show()
```
