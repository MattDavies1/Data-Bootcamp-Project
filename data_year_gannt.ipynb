```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
data_path = os.path.join("Resources", "data_tracker.csv")

data = pd.read_csv(data_path)
data
```

```python
#Estalish variables required for formatting Gantt

number_files = len(data)
y_ticks = np.arange(1,number_files+1)
files = list(data['Data File'])
```

```python
fig, gnt = plt.subplots()

# Format output
gnt.set_title('Data Set Overlap')
gnt.set_xlabel('Year') 
gnt.set_ylabel('Data Set') 
gnt.grid(True)

# Set axes
gnt.set_ylim(0,number_files + 1)
gnt.set_xlim(1960, 2020)

# Set and label ticks on y-axis 
gnt.set_yticks(y_ticks)
gnt.set_yticklabels(files) 

# Create empty variable to track location on graph in for loop
location = 0

# Loop to make Bars for Gantt
for index, row in data.iterrows():
    start = int(row['Start Date'])
    end = int(row['End Date'])
    gnt.broken_barh([(start, end-start)], (location+0.5, 0.9))
    location += 1
    
plt.savefig('Outputs/dataset_gantt.png', format = 'png', dpi = 300, bbox_inches='tight')
```
