```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
data_path = os.path.join('Resources', 'canada_population.csv')
ontario = pd.read_csv(data_path)
ontario = ontario[["When", "Ontario"]]
ontario["REF_DATE"] = ontario["When"].str[-4:]
ontario = ontario[["Ontario", "REF_DATE"]]
ontario["REF_DATE"] = ontario["REF_DATE"].astype(int)
ontario = ontario.drop_duplicates(subset=["REF_DATE"], keep='last')
ontario = ontario.loc[(ontario["REF_DATE"] >= 1998)&
                     (ontario["REF_DATE"] <= 2013), :]
ontario
```

```python
data_path = os.path.join('Resources', 'ontario_impaired_driving.csv')
data = pd.read_csv(data_path)
data_clean1 = data.loc[(data["Violations"]=="Total impaired driving [910]") &
                       (data["UOM"] == "Rate") &
                       (data["Statistics"]=="Rate per 100,000 population"),:]
impaired_driving_on = data_clean1[["REF_DATE", "GEO", "Violations", "SCALAR_FACTOR", "VALUE"]]
impaired_driving_on = impaired_driving_on.loc[impaired_driving_on["REF_DATE"] <= 2018, :]
impaired_driving_on
merged_driving = pd.merge(impaired_driving_on, ontario, on="REF_DATE")
merged_driving["Impaired Driving per 100,000"] = (merged_driving["VALUE"]/merged_driving["Ontario"]) * 100000
merged_driving
```

```python
data_path = os.path.join('Resources', 'cannabis_usage_1961_2018.csv')
data = pd.read_csv(data_path)
data_clean1 = data.loc[(data["Frequency of use"]=="Total, all frequencies") &
                       (data["UOM"] == "Persons"),:]
data_clean = data_clean1.groupby("REF_DATE").sum()
data_clean = data_clean[["VALUE"]]
data_clean.reset_index()
merged_cannabis = pd.merge(data_clean, ontario, on="REF_DATE")
merged_cannabis["Impaired Driving per 100,000"] = (merged_cannabis["VALUE"]/merged_cannabis["Ontario"]) * 100000
merged_cannabis
```
